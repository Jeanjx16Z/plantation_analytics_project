## Load Cleaned Dataset

```python
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_data.csv")
df.head()
```

## Buat dim_region

```python
dim_region = df[["region"]].drop_duplicates().reset_index(drop=True)
dim_region["region_id"] = dim_region.index + 1
dim_region
```

## Merge region_id ke fact

```python
df = df.merge(dim_region, on="region")
```

## Buat dim_date

```python
df["date"]= pd.to_datetime(df["date"])

dim_date = df[["date"]].drop_duplicates().reset_index(drop=True)
dim_date["date_id"] = dim_date.index + 1
dim_date["month"] = dim_date["date"].dt.month
dim_date["year"] = dim_date["date"].dt.year
dim_date
```

## Merge ke fact

```python
df = df.merge(dim_date, on="date")
```

## Final fact table

```python
fact_production = df[[\
    "region_id",
    "date_id",
    "production_ton",
    "sales_revenue",
    "operational_cost",
    "profit"
]]

fact_production.head()
```

*Saya mengubah dataset flat menjadi star schema untuk meningkatkan efisiensi analytics dan mengurangi redundansi data. Fact table menyimpan metrik produksi, sementara dimension table menyimpan konteks waktu dan region.*
